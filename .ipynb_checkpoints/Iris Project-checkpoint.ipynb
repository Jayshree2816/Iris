Iris data deom Seaborn

```python
iris = sns.load_dataset('iris')
```

```python
iris.describe()
```

```python
iris.info()
```

```python
iris.groupby('species').size()
```

```python
iris.isnull().sum()
```

```python
sns.pairplot(iris, hue='species',height=3, aspect =1)
```

```python
iris.hist(edgecolor ='black', linewidth = 1.2, figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.subplot(2,2,1)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species', y='petal_length', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species', y='petal_width', data=iris)
```

```python
iris.boxplot(by='species',figsize=(12,8))
```

```python
pd.plotting.scatter_matrix(iris, figsize=(12,10))
```
